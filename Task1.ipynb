**frequency distribution of words in a text**

```
textsample1='Experts on constitutional and political affairs say sovereignty vested in the people, limited government, the principle of separation of powers, independent judiciary, human rights and fundamental rights and inclusive governance system are the components of constitutionalism. The breach of any of these components means an attack on constitutionalism.'

fd=nltk.FreqDist(textsample1.split())
fd
```

**Conditional frequency distribution of words in a text**

```
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in textsample1.split())
cfd[4]
```
